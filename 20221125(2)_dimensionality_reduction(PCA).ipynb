## Loading a dataset

```python
import pandas as pd
iris = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data', names=['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'label'])
iris
```

```python
X = iris.drop(['label'], axis=1)
#label은 없어도 됨. 하나의 변수만으로 PCA를 때리는 것.
X
```

## Principal Component Analysis (PCA)

```python
from sklearn.decomposition import PCA
model = PCA(n_components=4)  # n_components specifies the new dimensionality
#이 경우는 4개 다 쓰는 상황
X_reduced = model.fit(X).transform(X) # or just pca.fit_transform(X)
#어쨌든 위에 X-reduced가 출력값이라고 할 수 있음.
pd.DataFrame(X_reduced) #df로 만들어서 보기 좋게 한 것뿐
```

위의 숫자들을 이해하자.

첫번째 칼럼에 나오는 것은 오리지널 축이 이동하고 로테이트해서 가장 두꺼운 축에 해당하는 값들.
값들이 당연히 달라질 수 밖에 없음.
축이 이동했으니까.

아래에 나오는 행렬 개념 보충설명

행렬은 직사각형이다. 행렬의 직사각형의 숫자들이 있으면 (3x2)이라고 하면 행이 3개,열이 2개

행렬은 곱하기가 가능함.
3x2의 행렬은 2xㅁ와만 곱하기 가능함.
그 이유는 짝을 맞추어서 곱해야 하므로 (노타빌리티 정리함) 

```python
#모델은 이미 훈련되어 데이터에 맞게 fitting하면서 중심으로 들어가서 왔다갔다해서 찾아냄.
#그 정보가 아래에 들어있음.
print('Components:', model.components_)
#여기서 나오는 정보가 매우 중요함.행렬값이 나옴.
#원래 4차원 벡터가 150개 있었음. #pca 했더니 똑같이 4차원의 150개
#(150x4)이 (150x4)으로 되었는데, 즉 CPA후상태에 무엇을 곱하면,원래처럼 다시 나올지에 대한 정보가 들어있는 행렬임.
#지금 여기 나온 행렬이 (4x4)인데 (150x4)와 (4x4)를 곱하면 (150x4) 나오니까~
#원상복구할 때 필요한 템플릿 행렬이 나온 것.
print('Explained Variance:', model.explained_variance_)
#standard deviation(표준편차)를 제곱한 varience(분산)임.
#얼마나 데이터의 두께를 커버하느냐에 대한 것.뒤에 두개를 봤더니 너무 미약하네 앞 두개만 써야지 할 수도 있음.
print('Explained Variance Ratio:', model.explained_variance_ratio_ )
#이것은 분산을 백분위로 나타내서, 각각 이 데이터의 몇 퍼센트를 설명하는지가 나와있음.
#첫번째 축은 0.92... 이런 식으로 #다 합치면 1임.
```

```python
from sklearn.decomposition import PCA
model = PCA(n_components=2)  # n_components specifies the new dimensionality
#앞서 4개의 설명력을 살펴보니 2개만 봐도 될 거 같아서 여기서는 2개만 쓰겠다고 했지만, 항상 그럴 필요는 없음,
#4개 다 쓰겠다고 해도 됨.#그러나 5라고 하면 안된다. (MAX가 오리지널 차원이다)
X_reduced = model.fit(X).transform(X) # or just pca.fit_transform(X)
pd.DataFrame(X_reduced)
```

```python
print('Components:', model.components_)
#그럼 원상복귀시키는 행렬이 달라짐, (2,4)가 나옴.
#PCA 후인 (150X2)에 (2X4) 곱하면 (150X4) 나오므로
print('Explained Variance:', model.explained_variance_)
# Variance는 설명력, 두께로 이해하면 된다.
print('Explained Variance Ratio:', model.explained_variance_ratio_ )
```

```python
X_approx = model.inverse_transform(X_reduced) #원상복귀시켜주는 과정
#X_reduced가 PCA 후의 행렬임. model 안에 이미 원복 행렬이 있기에 inverse_transform라는 함수 써서 원복시키기
#오리지널을 X_approx라고 써놨는데 사실 이건 완벽 오리지널이 될 수 없음. 
#우리가 component를 2개만 가져온 상황이므로 (만약 4개하면 완벽 원복도 되겠지만)
pd.DataFrame(X_approx)
```

```python
X #이게 찐 원복임
#위에 것과 비교해보면 조금 차이는 있지만 거의 비슷함.
#원래는 150X4나 있었는데, PCA해서 원복을 했다는 것은 그만큼 필요가 없다는 걸 증명한 것.
#모두 들고 있을 필요없다. PCA된 값만 들고 있어도, 그 행렬만 있어도 원복을 할 수 있다.
```

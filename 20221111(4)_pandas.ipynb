### Constructing a DataFrame using dict

csv로 해서 가져올 수도 있지만, dict로 만들어서 한땀 한땀 만들 수도 있음.

```python
# Some input data - a small sample of the iris dataset
data = {'sepal_length': [6.9, 6.9, 4.8, 5.4, 4.6],
        'sepal_width': [3.2, 3.1, 3.4, 3.0, 3.6],
        'petal_length': [5.7, 5.1, 1.9, 4.5, 1.0],
        'petal_width': [2.3, 2.3, 0.2, 1.5, 0.2],
        'species': ['virginica', 'virginica', 'setosa', 'versicolor', 'setosa']}
#전형적인 dict형태인 위 data, key와 value값
#key가 나중에 칼럼명 value는 그 칼럼 아래에 쭉 나오는 값
#열이 5개
df = pd.DataFrame(data) #dict를 데이터프레임으로 만들어주는 것
df 
```

### Constructing a DataFrame using pd.Series

pd.Series로 value값만 따로 해주는 방식
방식이 다를 뿐
dict를 이렇게도 가능
미리 만들고 변수로 받는 느낌.

```python
sepal_length = pd.Series([6.9, 6.9, 4.8, 5.4, 4.6])
sepal_width = pd.Series([3.2, 3.1, 3.4, 3.0, 3.6])
petal_length = pd.Series([5.7, 5.1, 1.9, 4.5, 1.0])
petal_width = pd.Series([2.3, 2.3, 0.2, 1.5, 0.2])
species = pd.Series(['virginica', 'virginica', 'setosa', 'versicolor', 'setosa'])

df = pd.DataFrame({'sepal_length': sepal_length, 'sepal_width': sepal_width, 'petal_length': petal_length, 'petal_width': petal_width, 'SPECIES': species})
df
```

### Changing column names

이것은 젤 위에 있는 칼럼명 바꿔도 보는 것.

```python
# Change the columns #위에는 species가 대문자인데 여기서 소문자로 변경
df.columns = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'species']
df
```

### Accessing columns

한 셀만 가져올 수도 있고, 조건을 걸어서 10이상만 가져오라고 할 수도 있고 등등

```python
# Access one column #칼럼 정보 가져오기_방법1 
df.loc[:, 'sepal_length'] #행, 열 정보를 모두 적은 것, ;는 all이라는 뜻
```

```python
# Access one column (short-hand) #칼럼 정보 가져오기_방법2 (간단)
df['sepal_length'] #단순히 
```

```python
# Access one column (shortest-hand) #칼럼 정보 가져오기_방법3 (더 간단)
df.sepal_length
```

```python
# Access two columns  #칼럼 두개 정보 가져오기_방법1
df.loc[:, ['sepal_length', 'sepal_width']] #리스트로 두개정보 가져온 것
```

```python
# Access two columns (short-hand) #칼럼 두개 정보 가져오기_방법2 (간단)
df[['sepal_length', 'sepal_width']] 
```

### Accessing rows

```python
# Access one row 
df.loc[3, :] #3번째 로우에 모든 값들 가져오기
```

```python
# Access two rows
df.loc[[3, 0], :] #3번째랑 0번째 
```

```python
# Access a range of rows
df.loc[2:5, :] #이건 2,3,4 번째 row정보 다 가져오는 것 (5는 불포함)
```

```python
# Filter keeping only rows with `sepal_length` > 5
df[df.sepal_length > 5] #sepal length가 5보다 큰 아이들만 가져와라
```

### Accessing a cell

이제는 셀별로

```python
df.loc[:, 'sepal_length'][4] #한칼럼 통째로 가져오고 거기서 4번째꺼
```

```python
df['sepal_length'][4]
```

```python
df.sepal_length[4]
```

### Adding/Removing new columns

```python
df['new_column'] = 1 #새로운 이름인 new column 만들고 1의 값 다 넣음.
df
```

```python
df['new_column'] = [1, 2, 3, 4, 5] 
#새로운 이름인 new column 만들고 리스트로 해서 각각의 값 다 넣음.
#모든 값이 1이었는데 replace하는 느낌
df
```

```python
df = df.drop('new_column', axis=1) # axis=1: column #axis1 자체가 칼럼얘기
df
```

### Adding/Removing new rows

```python
df.loc[5, :] = '' #loc 함수 활용해서 5번째 row에 대한 것을 ''(없애라는 뜻)
df
```

```python
df = df.drop(5, axis=0) # axis=0: row #row를 없애라
df
```
